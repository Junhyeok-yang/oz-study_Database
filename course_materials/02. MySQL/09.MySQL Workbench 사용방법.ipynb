### 데이터 베이스 생성

**1. 데이터베이스 생성**

```sql
-- mydatabase라는 이름의 데이터베이스 생성
CREATE DATABASE mydatabase;
```

**2. 데이터베이스 목록 조회**

```sql
-- 모든 데이터베이스 목록 조회
SHOW DATABASES;
```

**3. 데이터베이스 사용**

```sql
-- mydatabase 데이터베이스 사용
USE mydatabase;
```

**4. 데이터베이스 삭제 (IF EXISTS 옵션 사용)**

```sql
-- mydatabase 데이터베이스 삭제 (만약 존재하면)
DROP DATABASE IF EXISTS mydatabase;
```
