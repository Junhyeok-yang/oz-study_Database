# 21.1) 실습 - 간단한 스키마 설계하기

**(1) user 테이블 만들기**

- id: 자동으로 생성됩니다. (PK)
- password: 4자리 랜덤
- name: 한글3자리
- gender: male, female 중 랜덤선택
- email: 5자리영문@gmail.com
- birthday: 6자리 숫자 랜덤하게
- age: 2자리 숫자 랜덤
- company: [samsung, lg, hyundai]
- **해설**
    
    ```sql
    CREATE TABLE user (
        id INT AUTO_INCREMENT PRIMARY KEY,
        password VARCHAR(4),
        name VARCHAR(3),
        gender ENUM('male', 'female'),
        email VARCHAR(15),
        birthday CHAR(6),
        age TINYINT,
        company ENUM('samsung', 'lg', 'hyundai')
    );
    ```
    
    - ENUM: enumeration
        - **`ENUM`** 타입의 필드는 사전에 정의된 문자열 값 목록 중 하나의 값을 가질 수 있습니다. 이러한 특성 덕분에 **`ENUM`**은 특정 범위의 값만을 허용해야 하는 경우에 유용

**(2) board 테이블 만들기**

- id
- title: 5글자
- content: 10글자
- likes: 1~100사이 숫자
- img: “c”
- created: 오늘 날짜
- user_id (foreign_key)
- **해설**
    
    ```sql
    CREATE TABLE board (
        id INT AUTO_INCREMENT PRIMARY KEY,
        title VARCHAR(5),
        content VARCHAR(10),
        likes INT CHECK (likes BETWEEN 1 AND 100),
        img CHAR(1) DEFAULT 'c',
        created DATE DEFAULT CURRENT_DATE,
        user_id INT,
        FOREIGN KEY (user_id) REFERENCES user(id)
    );
    ```
